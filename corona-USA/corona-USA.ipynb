```python
import numpy as np
import pandas as pd 
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv("counties.csv")
df.head()
```

```python
total = df.groupby(['date'], axis=0,as_index=False).sum()
total['date'] = pd.to_datetime(total['date'], format='%Y-%m-%d')
total['Day of Year 2020']=total['date'].dt.dayofyear
total_days = total.groupby(['Day of Year 2020'], axis=0).sum()
total_days
```

```python

for day in range (2,total_days.shape[0]):
    total_days[['cases','deaths']].iloc[:day].plot(marker='x', logy=True, grid=True, xlim=[21,90])
    plt.savefig("images/"+str(day).zfill(4)+".png")
    plt.close()
```
